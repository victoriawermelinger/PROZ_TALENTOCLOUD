**1.1 O que é HTML (história e definição)**

Hypertext Markup Language (HTML), em português, Linguagem de Marcação de Hipertexto, é um conjunto de sinais e códigos aplicados a um texto, ou dados, que contribui para definir exibições na tela e de estruturas de dados em um sistema.

image.png

Muitos acham que a HTML é uma linguagem de programação, mas ela não é, pois não possui propriedades como operadores lógicos, funções, variáveis etc.

A Hypertext Markup Language serve para o navegador interpretar os elementos de uma página. É ela que dá a estrutura da página, ou seja, dá significado e organiza as informações de uma página na web. Sem a HTML, o navegador não conseguiria exibir textos, elementos ou carregar imagens.

**Principais tipos de tags**

É possível categorizar as tags em dois tipos: o block-level (nível de bloco) e o inline (em linha). O primeiro ocupa todo o espaço do seu elemento pai e o segundo ocupa apenas o espaço do seu conteúdo, como no exemplo ilustrado.
