1) Na variável está discriminada que a nota é IGUAL a 10. Já na sentença, está discriminada que
a nota é menor OU igual a 10. Portanto, o resultado final dessa sentença é VERDADEIRO.
2) Na variável, é apresentada “nota” IGUAL a 6 e “faltas” IGUAL a 4. Na sentença está descrito
que “nota” é MENOR ou IGUAL a 6 e isso está correto. “Faltas” está descrita como MENOR
ou IGUAL a 3 e isso está incorreto. Como na sentença é utilizado o operador lógico de
CONJUNÇÃO (AND), pela tabela verdade para que o resultado final seja verdadeiro, é
necessário que todos os valores declarados sejam verdadeiros. Portanto, o resultado final
desse exercício é FALSO.
3) Na declaração de variáveis há “convidados” IGUAL a 3 e “fumante” IGUAL a false. Na
sentença, está descrito como “convidados” MAIOR QUE 4 (isso contradiz a variável) e
“fumante” IGUAL a true (isso contradiz a variável). A expressão usa o operador lógico "OR",
que retorna Verdadeiro se pelo menos uma das condições da sentença for verdadeira. No
entanto, neste caso, ambas as partes da expressão são falsas. Portanto, o resultado final da
sentença é FALSO.
4) Na variável é apresentado “dia” IGUAL a 4. Na sentença são apresentadas duas opções que
fazem uma disjunção (OR), sendo elas “dia” IGUAL a sab e “dia” IGUAL a dom. Nenhuma das
opções é verdadeira, portanto, o resultado final é FALSO.
5) Na declaração apresentada diz que “feriado” é IGUAL a true. Na sentença, há o operador
lógico de negação (not) para o que é apresentado logo a seguir, que é “feriado” == false.
Sendo negativo o false, então a sentença informa que é true. Portanto, o resultado final é
VERDADEIRO.
6) Na variável são apresentadas duas opções: dia IGUAL a “ter” e feriado IGUAL a false. Na
sentença, essas duas variáveis foram declaradas como: dia IGUAL a “seg” (falso) e feriado
IGUAL a false (verdade). Porém, antes da sentença de feriado há o ponto de exclamação (!)
que, assim como foi apresentado na aula de tipos de operadores, significa NEGAÇÃO. Então,
é falso negativo o feriado, passando a ser o feriado == true. Portanto, o resultado final dessa
sentença é FALSO.
